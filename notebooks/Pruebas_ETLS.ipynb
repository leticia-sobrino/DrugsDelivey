```python
import pyproj
_projections = {}

def unproject(z, l, x, y):
    if z not in _projections:
        _projections[z] = pyproj.Proj(proj='utm', zone=z, ellps='WGS84')
    if l < 'N':
        y -= 10000000
    lng, lat = _projections[z](x, y, inverse=True)
    return (lat, lng)
```

```python
x = list(farmacias_clean.localizacion_coordenada_x)
y = list(farmacias_clean.localizacion_coordenada_y)
```

```python
lat, lng = unproject(30, "N", x, y)
```

```python
farmacias_clean['longitud'] = lng
```
